# Local Variable:
Which execute with in the function. we can not access the local variable outside the function.

```python
def test1(n):
    l = []
    for i in range(1, n+1):
        l.append(i)
    return l
```

```python
test1(5)
```

```python
l  # l is local variable so we cann't access it outside the function
```

# Global Variable:
WHich execute with in the fuction and also we can access it outside the function..

```python
a = 'hello'

def simple():
    b = 'World'
    print(a,b)
print(a)
```

```python
simple()
```

```python
result = 10

def test3(num):
    global result
    result = num * num
test3(3)
print(result)
```

```python
scores = 0

def match():
    global scores
    scores = scores + 2
match()
print(scores)
```

```python
# Sum of n Natural Numbers with function using global variable...
scores = 0

def matchs(num):
    global scores
    for i in range(1, num+1):
        scores = scores + i
matchs(5)
print(scores)
```

```python
def test7(*args):
    l =[]
    for i in args:
        if type(i) == int:  # int values 
            l.append(i)  # add the int values into of i in the l..
        elif type(i) == list:  # here if i = list..
            for j in i: 
                if type(j) == int: # here it's check the int type's values
                    l.append(j) 
    return l
```

```python
test7(1,2,30,40,'Hello','Yes',[30,True,45,30.78],90,True)
```

```python
def find_int1(*args):
    l = []

    for i in args:

        if type(i) == int:
            l.append(i)

        elif type(i) == list:
            for j in i:

                if type(j) == int:
                    l.append(j)

                elif type(j) == list:
                    for k in j:
                        if type(k) == int:
                            l.append(k)

        elif type(i) == tuple:
            for j in i:
                if type(j) == int:
                    l.append(j)

    return l
```

```python
# 1,2.5,True,[10,15.2,False], (12,12.6,'Python'), [30,50,10.5,[1,2,3,4,'Hello',40],46.6,39],29


def find_int(*args):
    l = []
    for i in args:
        if type(i) == int:
            l.append(i)
        elif type(i) == list:
            for j in i:
                if type(j) == int:
                    l.append(j)
                elif type(j) == list:
                    for y in j:
                        if type(y) == int:
                            l.append(y)
        elif type(i) == tuple:
            for j in i:
                if type(j) == int:
                    l.append(j)
    return l
```
