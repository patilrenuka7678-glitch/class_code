# Lambda Function

```python
add = lambda a, b: a + b
```

```python
add(3,4)
```

```python
sqaure = lambda i : i * i
```

```python
sqaure(4)
```

```python
compare = lambda a, b : print(a, 'is greater') if a > b else print(b,'is greater')
```

```python
compare(74,50)
```

```python
odd_even = lambda a : print(a, 'Even') if a % 2 == 0 else print(a,'Odd') 
```

```python
odd_even(98)
```

```python
odd_even(67)
```

```python
length = lambda a : len(a)
```

```python
length('renuka')
```

```python
upper = lambda text : print(text)
a = print(upper('renukA Sambhaji patil'))
a
```

# comprehension

```python
# List comprehension 

numbers = [i for i in range(1,6)]
```

```python
print(numbers)
```

```python
numbers1 = [i for i in range(1,11) if i % 2 == 0]
```

```python
print(numbers1)
```

```python
# syntax = variable = [expression for item in interable]
square = [i * i for i in range(1,6)]
print(square)
```

```python
# set is unorderd data structure so there for it return the output like this....
# we use sorted 
square_set = {i * i for i in range(1,11)}
print(sorted(square_set))
```

```python
# Tuple is generator expression thier for it's return the generator object.....
# tuple add in starting to get the list.....

number = (i for i in range(1,6))
print(number)
```

```python
# tuple add in starting to get the list.....

number = tuple(i for i in range(1,6))
print(number)
```
