# Set 

```python
s = {10,20,30,40,50}
s
```

```python
s.add(60)
s
```

```python
s.remove(60) # remove only delete that element which exist into the set nither it's give the key error.
s
```

```python
s.discard(60)   # discard also used for the delete element from the set but it not throw the keyerror like .remove().
s
```

```python
a = {1,2,3,4}  
b = {3,4,5,6}
a.union(b)
```

```python
a.intersection(b) # return the elements which in the both set.
```

```python
b.difference(a)
```

```python
a.difference(b)
```
