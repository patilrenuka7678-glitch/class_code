# While Loop

```python
age = 15

while age < 18:
    print('Student',age)
    age = age + 1
```

```python
button = 'hello'

while button != 'Exit':   # hello != exit  here condition will be a true 
    print(button)
    button = input('Enter the name of button')   # here when we write Exit so condition becomes false.
```

# Statements

# 1) Break

```python
for i in range(1,11):
    if i == 5:
        break 
    else:
        print(i)
```

# 2) Continue

```python
for i in range(1,11):
    if i == 5:
        continue
    else:
        print(i)
```

# 3) pass

```python
for i in range(1,11):
    if i == 1:
        pass
    elif i == 6:
        pass
    else:
        print('You will get panishment',i)
```
